```python
import win32com.client
```

- 워크북/시트 불러오기

```python
wb = excel.Workbooks.Open('D:/작업방/python_RPA/sample.xlsx')
```

```python
ws = wb.Worksheets('sheet1')
```

- 데이터 행 개수 세기

```python
i=1

cell_value = ws.Cells(i,1).Value

while cell_value != None:
    i+=1
    cell_value = ws.Cells(i,1).Value

max_row = i-1
```

```python
i=1

# cell_value = ws.Cells(i,1).Value

while ws.Cells(i,1).Value != None:
    i+=1
    # cell_value = ws.Cells(i,1).Value

max_row = i-1
```

```python
max_row
```

- 합계 제목

```python
ws.Cells(max_row+1, 1).Value = '합계'
```

- 합계 계산

```python
import string

abc = string.ascii_lowercase

#  =SUM(B2:B11)

for col in range(2,6):
    ws.Cells(max_row+1,col).Formula = '=sum({}2:{}{}'.format(abc[col-1],abc[col-1],max_row)
```

```python
j=1

while ws.Cells(1,j).Value != None:
    j+=1
max_col = j-1
```

```python
max_col
```

```python
ws.Cells(1,max_col+1).Value = '평균'
```

```python
# =AVERAGE(B2:E2)

for row in range(2,max_row+2):
    ws.Cells(row, max_col+1).Formula = '=average(b{}:{}{}'.format(row,abc[max_col-1],row)
```
